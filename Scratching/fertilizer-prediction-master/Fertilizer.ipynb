```python
fert_name = {'Urea': 1,'DAP': 2,'14-35-14':3, '28-28':4, '17-17-17':5, '20-20':6,
       '10-26-26':7}
fert['Fertilizer Name'] = [fert_name[item] for item in fert['Fertilizer Name']]
```

```python
fert
```

```python
data_dum = pd.get_dummies(fert)
data_dum[:5]
```

```python
x = data_dum.drop("Fertilizer Name",axis=1)
y = data_dum[["Fertilizer Name"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=42)
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
x_train[:5]
```
